## De Otoño a Ártico en 24 Horas: Diagnóstico de un Frente Frío Extremo en Boulder

El 10 de noviembre de 2014 un frente frío intenso cruzó por la ciudad de Boulder, Colorado, transformando en cuestión de horas un típico paisaje otoñal en uno propio de latitudes casi polares. Aunque para quienes vivimos en regiones más cálidas este cambio puede parecer difícil de imaginar, la serie de fotografías de la Figura 1 muestra claramente cómo el campus de la Universidad de Colorado en Boulder pasó de condiciones otoñales relativamente templadas a un entorno con nieve, temperaturas bajo cero y sensación térmica plenamente invernal.

Este frente frío no fue un evento aislado a escala local, sino parte de una irrupción de aire ártico de gran extensión espacial que afectó amplias regiones de Estados Unidos al inicio de la temporada invernal 2014 2015. La configuración sinóptica estuvo asociada a una fuerte amplificación de la corriente en chorro, inducida días antes por una ciclogénesis extratropical muy intensa en el Mar de Bering. Esta perturbación favoreció un patrón de bloqueo que permitió que una masa de aire ártico descendiera de forma anómala hacia latitudes medias, alcanzando las Montañas Rocosas y las Grandes Planicies.

En el caso específico de Colorado, el paso del frente frío el 10 de noviembre produjo un descenso térmico abrupto en menos de 24 horas. En sectores cercanos a Boulder y Denver, las temperaturas máximas posteriores al frente no superaron aproximadamente −9 °C, con caídas cercanas a 25 °C respecto al día previo. Además del enfriamiento marcado, el evento estuvo acompañado de vientos intensos y episodios de nieve, consolidando un escenario típicamente invernal mucho antes de lo habitual para mediados de noviembre.

Desde el punto de vista dinámico, este caso constituye un ejemplo didáctico de cómo las interacciones a escala planetaria, en particular las modificaciones en el patrón ondulatorio de la corriente en chorro, pueden traducirse en impactos locales muy marcados. En este ejercicio analizaremos datos atmosféricos asociados a este evento con el objetivo de comprender la evolución temporal del frente frío y los cambios en variables como temperatura, viento y presión durante su paso sobre Boulder.

## Estación meteorológica

En el techo del edificio de física *Duane Physics Building* se encuentra instalada una estación meteorológica automática.

<img src="PAOSwxstation1.jpg" width="400">

Las coordenadas geográficas de la estación son:

- **Latitud:** 40.0078° N  
- **Longitud:** 105.2683° W  
- **Elevación:** 1660 m sobre el nivel del mar  

La estación corresponde a un sistema **Davis Instruments Wireless Vantage Pro2**, ampliamente utilizado en monitoreo meteorológico de superficie.

En el archivo `wxobs20141110.txt` se presentan los valores de temperatura del aire en °F (y otras variables) registrados cada 5 minutos por esta estación, comenzando a las 00:00 (hora local) del día 10 de noviembre de 2014. Estos datos permitirán analizar la evolución térmica asociada al paso del frente frío sobre Boulder.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Dada la estructura del archivo, por simplicidad asignaré los nombres de las columnas manualmente.
columns = [
    "Date", "Time",
    "OutTemp_F", "HiOutTemp_F", "LowOutTemp_F", "OutHum_pct", "DewPt_F",
    "WindSpeed", "WindDir", "WindRun",
    "HiWindSpeed", "HiWindDir",
    "WindChill_F", "HeatIndex_F", "THWIndex_F", "THSWIndex_F",
    "Bar",
    "Rain", "RainRate",
    "SolarRad", "SolarEnergy", "HiSolarRad",
    "HeatDD", "CoolDD",
    "InTemp_F", "InHum_pct", "InDew_F", "InHeat_F",
    "InEMC", "InAirDensity",
    "ET", "WindSamp", "WindTx", "ISSRecept", "ArcInt"
]

# Leer ignorando las primeras 3 líneas
df = pd.read_csv('wxobs20141110.txt', sep=r"\s+", skiprows=3, header=None, names=columns, engine="python")

# Crear columna datetime combinando Date y Time (formato 12h con a/p)
df["datetime"] = pd.to_datetime(df["Date"] + " " + df["Time"].str.replace("a", " AM", regex=False).str.replace("p", " PM", regex=False), format="%m/%d/%y %I:%M %p")

# Convertir de Fahrenheit a Celsius
df["OutTemp_C"] = (df["OutTemp_F"] - 32) / 1.8

# Usar como índice
df = df.set_index("datetime").sort_index()

# Voy a recortar los datos para conservar solamente los del 10 de noviembre de 2014
start_day = pd.Timestamp("2014-11-10 00:00")
end_day   = pd.Timestamp("2014-11-11 00:00")
df = df.loc[start_day:end_day]

# Mostrar las primeras 15 filas
df.head(15)
```

```python
# Extraer únicamente la temperatura en °C en un dataframe
df_temp = df[["OutTemp_C"]].copy()

# Mostrar las primeras 15 filas
df_temp.head(15)
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

# Eje izquierdo: Temperatura
ax.plot(df_temp.index, df_temp["OutTemp_C"], linewidth=2, color="red")

ax.set_ylabel("Temperatura (°C)")

# Definir inicio y fin del día
start = pd.Timestamp("2014-11-10 00:00")
end   = pd.Timestamp("2014-11-11 00:00")

ax.set_xlim(start, end)

# Ticks cada 2 horas incluyendo 24
hours = pd.date_range(start=start, end=end, freq="2H")
ax.set_xticks(hours)
labels = [f"{h.hour:02d}" if h != end else "24" for h in hours]
ax.set_xticklabels(labels)

ax.set_xlabel("Hora")

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.grid(True, which="major", linewidth=0.6, alpha=0.25)

plt.tight_layout()
plt.show()
```

## Cálculo de la derivada temporal de la temperatura

Con el objetivo de cuantificar la tasa de cambio temporal de la temperatura durante el paso del frente frío, calcularemos la derivada primera $\frac{dT}{dt}$ utilizando una aproximación numérica de diferencias finitas centradas.

Si la serie de temperatura está muestreada en intervalos regulares de tiempo $\Delta t$, la derivada en un instante $t_0$ puede aproximarse como:

$$
f'(t_0) \approx \frac{f(t_0 + \Delta t) - f(t_0 - \Delta t)}{2\Delta t}
$$

Esta expresión corresponde al esquema de diferencias centradas de segundo orden, cuyo error de truncamiento es del orden $\mathcal{O}(\Delta t^2)$.

En nuestro caso:

- $f(t) \equiv T(t)$ representa la temperatura del aire.
- $\Delta t = 5$ minutos.
- La derivada resultante representa la tendencia instantánea de la temperatura.

El cálculo de $\frac{dT}{dt}$ permitirá identificar periodos de enfriamiento o calentamiento rápido, en particular el máximo enfriamiento asociado al paso frontal.


```python
# Paso temporal en segundos (5 minutos)
dt = 5 * 60


T = df_temp["OutTemp_C"].values


# Se debe tener en cuenta que como necesitamos un punto en t+1 y un punto en t-1, 
# no podremos calcular la derivada para el primer y último tiempo de nuestra serie

# Inicializar array con NaN
dTdt = np.full(len(T), np.nan) # El primer y último valor permanecerá como NaN al final del procedimiento 

# Diferencias centradas para puntos interiores
for i in range(1, len(T)-1):
    dTdt[i] = (T[i+1] - T[i-1]) / (2 * dt)

    
# Convertir a °C por hora
dTdt = dTdt * 3600

# Agregar al DataFrame
df_temp["dTdt_C_per_h"] = dTdt

df_temp.head(15)
```

```python
fig, (ax1, ax2) = plt.subplots(
    2, 1, figsize=(10, 6),
    sharex=True,
    gridspec_kw={"hspace": 0.15}
)

# Panel superior: Temperatura
ax1.plot(df_temp.index, df_temp["OutTemp_C"],
         linewidth=2, color="red")

ax1.set_ylabel("Temperatura (°C)")
ax1.set_xlim(start, end)

ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
ax1.grid(True, linewidth=0.6, alpha=0.25)

# Panel inferior: Derivada
ax2.plot(df_temp.index, df_temp["dTdt_C_per_h"],
         linewidth=2, color="blue")

ax2.axhline(0, color="black", linewidth=1.5, alpha=0.9)

ax2.set_ylabel("dT/dt (°C h⁻¹)")
ax2.set_xlabel("Hora")

ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)
ax2.grid(True, linewidth=0.6, alpha=0.25)

# Configuración eje x
ax2.set_xticks(hours)
ax2.set_xticklabels(labels)


plt.show()
```

## Identificación del paso del frente frío mediante la tendencia térmica

Con el objetivo de identificar de manera objetiva el paso del frente frío, utilizaremos la derivada temporal de la temperatura, $\frac{dT}{dt}$.

La derivada nos permite cuantificar la tasa de cambio instantánea de la temperatura y, por lo tanto, detectar:

- El **inicio del enfriamiento**, cuando $\frac{dT}{dt}$ toma valores negativos significativos de manera sostenida.
- El **momento de estabilización**, cuando $\frac{dT}{dt}$ regresa a valores cercanos a cero después del descenso abrupto.

Desde el punto de vista físico, el paso frontal se caracteriza por un gradiente térmico marcado, que se traduce en una tendencia negativa intensa de la temperatura. Posteriormente, una vez que la masa de aire frío se establece, la tendencia vuelve a valores pequeños en magnitud, indicando un nuevo régimen térmico más estable.

A partir de estos dos instantes estimados:

1. Se calculará el **promedio de temperatura antes del inicio del enfriamiento**.
2. Se calculará el **promedio de temperatura después de la estabilización**.
3. Se representará gráficamente una transición idealizada entre ambos estados, permitiendo visualizar de forma clara el salto térmico asociado al paso del frente frío.

Este procedimiento combina análisis dinámico (mediante $\frac{dT}{dt}$) con una interpretación física del cambio de masa de aire sobre Boulder.


```python
# Voy a definir algunos parametros. 
# Como esto es un ejercicio puramente de salón de clase voy a jugar arbitrariamente con estos números.
# en principio los he definido de una forma arbitraria.
# Sería interesante buscar argumentos mas rigurosos para definirlos.
# Si encuentran cosas sobre esto por favor me lo comparten.

# Voy a definir que consideraré un "enfriamiento fuerte"
neg_thresh = -2.0   # °C/h

# Voy a definir un valor de dT/dt que basicamente es cero o "muy pequeño"
zero_band  =  0.4   # °C/h

# Voy a definir cuantos pasos consecutivos se debe mantener el comportamiento
# para considera que no estoy ante un brinco aleatorio 
persist_n  =  4

# Extraigo los datos de T y dTdt
T = df_temp["OutTemp_C"].copy()
dTdt = df_temp["dTdt_C_per_h"].copy()

# Como se puede ver en la figura anterior, la derivada tiene mucho ruido
# voy a suavizarla un poco con una ventana movil, solo para que sea mas fácil el proceso
# voy a escoger una ventana de 3 pasos, pero pueden jugar con esto. A mayor numero de pasos
# mayor será el suavizado y podrían perder información.
dTdt_smooth = dTdt.rolling(window=3, center=True, min_periods=1).mean()


# Voy a estimar el inicio de la llegada del frente frio.
# Para esto considero que dT/dt < neg_thresh por persist_n pasos consecutivos
start_mask = (dTdt_smooth < neg_thresh).rolling(persist_n, min_periods=persist_n).sum() == persist_n
# primer instante que cumple persistencia
t_start = start_mask[start_mask].index[0]  


# Voy a estimar el punto en donde ya ha pasado el frente frio.
# Voy a considerar que la temperatura se ha estabilizado |dT/dt| < zero_band por persist_n pasos consecutivos, pero buscando despues de t_start
post = dTdt_smooth.loc[t_start:]
end_mask = (post.abs() < zero_band).rolling(persist_n, min_periods=persist_n).sum() == persist_n
# primer instante que cumple persistencia
t_end = end_mask[end_mask].index[0]


#Voy a calcular los promedios antes y despues del paso del frente frio
# antes del inicio
T_pre  = T.loc[:t_start].iloc[:-1] 
Tpre_mean  = T_pre.mean()

# desde el fin hasta el final
T_post = T.loc[t_end:]       
Tpost_mean = T_post.mean()

print("Promedio antes (°C):         ", round(Tpre_mean, 2))
print("Promedio después (°C):       ", round(Tpost_mean, 2))
```

```python
fig, (ax1, ax2) = plt.subplots(
    2, 1, figsize=(10, 6),
    sharex=True,
    gridspec_kw={"hspace": 0.15}
)

# Panel superior: Temperatura
ax1.plot(df_temp.index, df_temp["OutTemp_C"],
         linewidth=2, color="red")

ax1.set_ylabel("Temperatura (°C)")
ax1.set_xlim(start, end)

ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
ax1.grid(True, linewidth=0.6, alpha=0.25)


ax1.hlines(Tpre_mean, xmin=start_day, xmax=t_start, linewidth=1.0, ls='--', color="black")
ax1.hlines(Tpost_mean, xmin=t_end, xmax=end_day, linewidth=1.0, ls='--', color="black")
# Línea oblicua conectando (t_start, Tpre_mean) con (t_end, Tpost_mean)
ax1.plot([t_start, t_end], [Tpre_mean, Tpost_mean],  linewidth=1.0, ls='--', color="black")

# Panel inferior: Derivada
ax2.plot(df_temp.index, df_temp["dTdt_C_per_h"],
         linewidth=2, color="blue")

ax2.axhline(0, color="black", linewidth=1.5, alpha=0.9)

ax2.set_ylabel("dT/dt (°C h⁻¹)")
ax2.set_xlabel("Hora")

ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)
ax2.grid(True, linewidth=0.6, alpha=0.25)

# Configuración eje x
ax2.set_xticks(hours)
ax2.set_xticklabels(labels)


plt.show()
```
