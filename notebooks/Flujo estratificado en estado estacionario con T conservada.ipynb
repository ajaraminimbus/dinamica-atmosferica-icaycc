# Flujo estratificado en estado estacionario con T conservada

Imagina un flujo establemente estratificado y en estado estacionario en el cual la temperatura $T$ es conservada (no hay calentamiento diabático).

1. ¿Cuál es la relación entre la advección horizontal de $T$ y la velocidad vertical $w$?
2. Da una explicación física de esta relación.


## Solución

### Conservación material de la temperatura

Si $T$ es conservada por las parcelas:

$$
\frac{D T}{D t} = 0
$$

La derivada material se escribe como:

$$
\frac{D T}{D t} =
\frac{\partial T}{\partial t}
+ u \frac{\partial T}{\partial x}
+ v \frac{\partial T}{\partial y}
+ w \frac{\partial T}{\partial z}
$$

Reorganizando nuestra expresión para $\frac{\partial T}{\partial t}$:

$$
\frac{\partial T}{\partial t}
= \frac{D T}{D t}
-\left(u\frac{\partial T}{\partial x}
+v\frac{\partial T}{\partial y}
+w\frac{\partial T}{\partial z}\right)
$$

Como el flujo está en estado estacionario (No hay cambios en el tiempo):

$$
\frac{\partial T}{\partial t} = 0
$$

Entonces:

$$
-\underbrace{\left(u \frac{\partial T}{\partial x}
+ v \frac{\partial T}{\partial y}\right)}_{\text{advección horizontal}}
- w \frac{\partial T}{\partial z}
= 0
$$

Despejando para $w$:

$$
w \frac{\partial T}{\partial z} = -\underbrace{\left(u \frac{\partial T}{\partial x}
+ v \frac{\partial T}{\partial y}\right)}_{\text{advección horizontal}}
$$


$$
w =
\frac{
-\overbrace{\left(
u \frac{\partial T}{\partial x}
+ v \frac{\partial T}{\partial y}
\right)}^{\text{advección horizontal}}
}{
\frac{\partial T}{\partial z}
}
$$

### Interpretación física

En un ambiente establemente estratificado:

$$
\frac{\partial T}{\partial z} \neq 0
$$

Existe un gradiente vertical de temperatura.

En nuestra tropósfera este gradiente veritical es aproximadamente:
    
$$
\frac{\partial T}{\partial z}
\approx -6.5 \,\text{K km}^{-1}
$$

Dado que 
$$
\frac{D T}{D t} = 0
$$

Se impone una restricción muy particular al movimiento de las parcelas: La parcela se mueve sobre una superficie $T = \text{cte}$.

Si una parcela experimenta advección horizontal hacia una región más cálida, para mantener $DT/Dt = 0$ debe desplazarse verticalmente hacia una región donde la temperatura ambiental compense ese cambio.

En otras palabras:

> La advección horizontal debe ser exactamente compensada por el movimiento vertical a lo largo de superficies de temperatura constante.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Vamos a crear un modelo sencillo bidimencional (x,z) es decir v=0

# Parámetros
u = 10.0          # viento en x (m/s)
dTdz = -0.0065    # gradiente vertical típico (K/m)

# Rango de gradiente en x
dTdx = np.linspace(-5e-6, 5e-6, 500)

# Advección horizontal de temperatura
A_h = - (u * dTdx)

# Velocidad vertical
w = A_h / dTdz




plt.figure()
plt.plot(A_h, w)

plt.axhline(0)
plt.axvline(0)
plt.fill_between(A_h, w, 0, where=(A_h < 0), color="blue", alpha=0.3)
plt.fill_between(A_h, w, 0, where=(A_h > 0), color="red", alpha=0.3)

plt.xlabel("Advección horizontal de temperatura (K/s)")
plt.ylabel("Velocidad vertical w (m/s)")
plt.title("w en función de la advección horizontal")

plt.show()
```
